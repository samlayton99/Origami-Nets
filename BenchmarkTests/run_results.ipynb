```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters for the grid size
rows, cols = 4, 5

# Create the subplot grid
plt.figure(figsize=(15, 10))
for i in range(rows * cols):
    plt.subplot(rows, cols, i + 1)
    plt.plot(np.random.rand(10))  # Example plot

# Add adaptive column titles
column_titles = [f"Column {i+1}" for i in range(cols)]  # Define titles for each column
top_margin = 0.92  # Adjust this to control the vertical position of column titles
for i, col_title in enumerate(column_titles):
    # Calculate the horizontal position of each title
    x_pos = (i + 0.5) / cols  # Center titles in each column
    plt.figtext(x_pos, top_margin, col_title, ha='center', fontsize=14, weight='bold')

# Adjust the layout to leave space for the titles
plt.subplots_adjust(top=0.85)  # Adjust `top` to control overall layout spacing
plt.tight_layout()
plt.show()
```
